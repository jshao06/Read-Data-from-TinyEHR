#### Import Packages

```python
import tinyehr
import sqlite3 as sql3
import pandas as pd
```

#### Build a Local SQLite Database of TinyEHR (MIMIC Format)

```python
db_path_mimic = tinyehr.build_sqlite(format='tinyehr_mimic_format', output='C:\\tinyehr\\tinyehr_mimic_format.db')
```

#### Connect to the Database

```python
conn_mimic = sql3.connect(db_path_mimic)
conn_mimic.execute('select * from admissions limit 1').fetchall()
```

#### Load Data in as Pandas DataFrames

```python
query = 'select * from patients;'
patients = pd.read_sql(query, conn_mimic)
patients.head()
```

```python
# Some common parameters
# coerce_float: Attempt to force numbers into floats
# parse_dates: List of columns to parse as dates
# chunksize: Number of rows to include in each chunk

patients_generator = pd.read_sql(query, 
                          conn_mimic,
                          coerce_float=True,
                          parse_dates=['dod'],
                          chunksize=5)
for id, patient in enumerate(patients_generator):
    if id < 5:
        print(f'patient id: {id}'.format(id))
        display(patient)
```
